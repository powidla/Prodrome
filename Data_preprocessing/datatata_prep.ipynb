```python
import re
import csv
import pandas as pd

sql_file = 'modified_data.sql'
csv_file = 'timepoints.csv'

# Define the CSV header
csv_header = ['id', 'userId', 'date', 'time', 'data', 'filled', 'registrationDate']

# Initialize a list to store the data rows
data_rows = []

with open(sql_file, 'r', encoding='utf-8') as sql_file:
    sql_data = sql_file.read()

# Extract the values within the parentheses for each INSERT INTO statement
insert_values = re.findall(r'\((.*?)\)', sql_data, re.DOTALL)

# Process each set of values
for values in insert_values:
    # Remove any leading/trailing spaces and split the values by commas
    values = values.strip().split(', ')

    # Clean up and process each value
    cleaned_values = []
    for value in values:
        # Remove surrounding quotes and escape characters
        cleaned_value = re.sub(r"^['\"]|['\"]$", '', value.strip())
        cleaned_values.append(cleaned_value)

    # Append the cleaned values to the data rows
    data_rows.append(cleaned_values)

# Save the data to CSV file
with open(csv_file, 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(csv_header)
    writer.writerows(data_rows)

print(f'Successfully converted SQL file to CSV: {csv_file}')
```
